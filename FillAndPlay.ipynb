```python
def run_query(query):
    mydb = mysql.connector.connect(
    host = host,
    user = user,
    passwd = password,
    database = database
    )
    cursor = mydb.cursor()
    
    cursor.execute(query)
    mydb.commit()
    cursor.close()
    mydb.close()
```

```python
def fill_companies():
    def query(*args):
        values = (f"('{name}','{date}')" for (name, date) in args)
        return f"insert into Companies (Name, FoundationYear) values {', '.join(values)};"
    
    q = query(*((name, f"{year}{month}01") for (name, year, month) in COMPANIES))
    print(q)
    run_query(q)
```

```python
def fill_regions():
    def query(*args):
        values = (f"({int(i)},'{name}',{value})" for (i, name, value) in args)
        return f"insert into Regions (Id, Name, ExistingGasAmount) values {', '.join(values)};"
    
    regions = [(code, region,random.randint(2, 20)) for (code, region) in REGIONS]
    q = query(*regions)
    print(q)
    run_query(q)
```

```python
def fill_gas_towers():
    rcodes = [code for (code, _) in REGIONS]
    ccodes = [i + 1 for i in range(len(COMPANIES))]
    
    def data():
        for name in NAMES:
            yield (random.choice(rcodes),
                  random.choice(ccodes),
                  random.randint(1,6),
                  random.randint(1980, 2019),
                  name)
            
    def query(args):
        values = (f"({ri}, {ci}, {value}, '{date}0101','{name}')" for (ri,ci,value,date, name) in args)
        return f"insert into GasTowers (RegionId, CompanyId, GasProductionRate, BuildingDate, Name) values {', '.join(values)};"
    d = list(data())
    q = query(d)
    print(q)
    run_query(q)
```

```python
def fill_employees():
    ccodes = [i + 1 for i in range(len(COMPANIES))]
    tcodes = [i + 1 for i in range(len(NAMES))]
    
    def data():
        for name in FIOS:
            yield (name,
                  random.choice(ccodes),
                  random.choice(tcodes),
                  random.randint(22, 65))
            
    def query(args):
        values = (f"('{name}', {ci},{ti}, {age})" for (name, ci, ti, age) in args)
        return f"insert into Employees (Name, CompanyId, TowerId, Age) values {', '.join(values)};"
    d = list(data())
    q = query(d)
    print(q)
    run_query(q)
```
